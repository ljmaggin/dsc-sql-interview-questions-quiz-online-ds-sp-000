## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

When you group your data by a certain variable you will need to choose an aggregation function for another variable. For instance a table of students and the classes they are in, if you want to see the number of classes each student is taking you can group by student and COUNT() the classes. The COUNT() is the aggregation function.
